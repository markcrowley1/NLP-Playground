```python
# Import packages
import re
import numpy as np
import pandas as pd
```

```python
# Read in train, val and test datasets
with open("tiny_shakespeare/train.csv", 'r') as f:
    csvreader = pd.read_csv(f)
    train_data = csvreader["text"].values[0]

with open("tiny_shakespeare/validation.csv", 'r') as f:
    csvreader = pd.read_csv(f)
    val_data = csvreader["text"].values[0]

with open("tiny_shakespeare/test.csv", 'r') as f:
    csvreader = pd.read_csv(f)
    test_data = csvreader["text"].values[0]

data = train_data.lower()
    
```

```python
""" Creating a bigram model """
# First we create a set of vocab used in shakespeare dataset
pattern = r'\w+|[^\w\s]|\n'
words = re.findall(pattern, data)
vocab = list(set(words))

# Get a subset of punctuation vocab
punc_pattern = r'[^\w\s]'
punctuation = set(re.findall(punc_pattern, data))

print(data[:80])
print(words[:20])
print(len(vocab))
```

```python
# Building a probability table
df = pd.DataFrame(0, index=vocab, columns=vocab, dtype=int)
# First count up occurences of word_y following word_x
for i in range(len(words) - 1):
    curr = words[i]
    next = words[i+1]
    df.at[curr, next] += 1
# The find P(y|x)
dfP = df.div(df.sum(axis=1), axis=0)
dfP
```

```python
output = ""
curr = "romeo"

for i in range(10):
    if curr in punctuation:
        output += curr
    else:
        output += " " + curr

    row = dfP.loc[curr]
    conditional_probabilities = dict(row[row != 0])
    curr = np.random.choice(list(conditional_probabilities.keys()), p=list(conditional_probabilities.values()))

print(output)
```
